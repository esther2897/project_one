```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframes = []
for file in os.listdir("datasets"):
    path = os.path.join("datasets",file)
    frame = pd.read_csv(path)
    dataframes.append(frame)
```

```python
PIT_2017 = dataframes[1]
PIT_2017.head()
gender_seven = PIT_2017[["State","Overall Homeless - Female, 2017","Overall Homeless - Male, 2017","Overall Homeless - Transgender, 2017","Overall Homeless - Gender Non-Conforming, 2017"]]
gender_seven = gender_seven.replace(r'^\s*$', np.nan, regex=True)
gender_seven = gender_seven[["Overall Homeless - Female, 2017","Overall Homeless - Transgender, 2017","Overall Homeless - Male, 2017","Overall Homeless - Gender Non-Conforming, 2017"]].astype(float)
gender_seven = gender_seven.rename(columns={"Overall Homeless - Female, 2017": "Female", "Overall Homeless - Transgender, 2017":"Transgender", "Overall Homeless - Male, 2017": "Male", "Overall Homeless - Gender Non-Conforming, 2017": "Gender Non-Conforming"})
gender_seven["State"] = PIT_2017["State"]
gender_seven = gender_seven.set_index("State")
sum_seven = gender_seven.loc["Total", :]
gender_seven_no_total = gender_seven.drop("Total")
colors = ['#ff9999','#99ff99','#66b3ff','#ffcc99']
gen_sev = gender_seven_no_total.plot.bar(stacked=True, title = 'Homeless Population in 2017, By Gender', color = {'Female': '#ff9999', 'Male': '#66b3ff', 'Transgender': '#99ff99', 'Gender Non-Conforming': '#ffcc99'})
gen_sev.set_xlabel("State")
gen_sev.set_ylabel("Number of Homeless People")
plt.savefig("gender_trends_by_state_2017.png")
```

```python
years = ["2017", "2018", "2019", "2020"]
fem_through_years = [sum_seven["Overall Homeless - Female, 2017"],sum_eight["Overall Homeless - Female, 2018"], sum_nine["Overall Homeless - Female, 2019"], sum_twenty["Overall Homeless - Female, 2020"]]
men_through_years = [sum_seven["Overall Homeless - Male, 2017"],sum_eight["Overall Homeless - Male, 2018"], sum_nine["Overall Homeless - Male, 2019"], sum_twenty["Overall Homeless - Male, 2020"]]
trans_through_years = [sum_seven["Overall Homeless - Transgender, 2017"],sum_eight["Overall Homeless - Transgender, 2018"], sum_nine["Overall Homeless - Transgender, 2019"], sum_twenty["Overall Homeless - Transgender, 2020"]]
nonconform_through_years = [sum_seven["Overall Homeless - Gender Non-Conforming, 2017"],sum_eight["Overall Homeless - Gender Non-Conforming, 2018"], sum_nine["Overall Homeless - Gender Non-Conforming, 2019"], sum_twenty["Overall Homeless - Gender Non-Conforming, 2020"]]
gender_total_throughout_years = pd.DataFrame({'Female':fem_through_years, 'Male':men_through_years, 'Transgender':trans_through_years, 'Gender Non-Conforming': nonconform_through_years}, index=years)
gender_total_female_male = gender_total_throughout_years[["Female", "Male"]]
gender_total_f_m = gender_total_female_male.plot.line(color = {'Female': '#ff9999', 'Male': '#66b3ff'}, title = "Homeless Population from 2017 to 2020")
gender_total_f_m.set_xlabel("Years")
gender_total_f_m.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
gender_total_f_m.set_ylabel("Number of Homeless People")
gender_total_f_m.grid()
plt.savefig("total_trend_female_male.png")
```

```python
gender_total_trans_non = gender_total_throughout_years[["Transgender", "Gender Non-Conforming"]]
gender_total_t_g = gender_total_trans_non.plot.line(color = {'Transgender': '#99ff99', 'Gender Non-Conforming': '#ffcc99'}, title = "Homeless Population from 2017 to 2020")
gender_total_t_g.set_xlabel("Years")
gender_total_t_g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
gender_total_t_g.set_ylabel("Number of Homeless People")
gender_total_t_g.grid()
plt.savefig("total_trend_trans_gen_non_con.png")
```
