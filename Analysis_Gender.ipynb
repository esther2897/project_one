```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframes = []
for file in os.listdir("datasets"):
    path = os.path.join("datasets",file)
    frame = pd.read_csv(path)
    dataframes.append(frame)
```

```python
PIT_2017 = dataframes[1]
PIT_2017.head()
gender_seven = PIT_2017[["State","Overall Homeless - Female, 2017","Overall Homeless - Male, 2017","Overall Homeless - Transgender, 2017","Overall Homeless - Gender Non-Conforming, 2017"]]
gender_seven = gender_seven.replace(r'^\s*$', np.nan, regex=True)
gender_seven = gender_seven[["Overall Homeless - Female, 2017","Overall Homeless - Transgender, 2017","Overall Homeless - Male, 2017","Overall Homeless - Gender Non-Conforming, 2017"]].astype(float)
gender_seven["State"] = PIT_2017["State"]
gender_seven = gender_seven.set_index("State")
sum_seven = gender_seven.loc["Total", :]
gender_seven_no_total = gender_seven.drop("Total")
gen_sev = gender_seven_no_total.plot.bar(stacked=True, title = 'Homeless Population in 2017, By Gender')
gen_sev.set_xlabel("State")
gen_sev.set_ylabel("Number of Homeless People")
```

```python
PIT_2018 = dataframes[2]
PIT_2018.head()
gender_eight = PIT_2018[["State","Overall Homeless - Female, 2018","Overall Homeless - Male, 2018","Overall Homeless - Transgender, 2018","Overall Homeless - Gender Non-Conforming, 2018"]]
gender_eight = gender_eight.replace(r'^\s*$', np.nan, regex=True)
gender_eight = gender_eight[["Overall Homeless - Female, 2018","Overall Homeless - Transgender, 2018","Overall Homeless - Male, 2018","Overall Homeless - Gender Non-Conforming, 2018"]].astype(float)
gender_eight.head()
gender_eight["State"] = PIT_2018["State"]
gender_eight = gender_eight.set_index("State")
sum_eight = gender_eight.loc["Total", :]
gender_eight_no_total = gender_eight.drop("Total")
gen_eig = gender_eight_no_total.plot.bar(stacked=True, title = 'Homeless Population in 2018, By Gender')
gen_eig.set_xlabel("State")
gen_eig.set_ylabel("Number of Homeless People")
```

```python
PIT_2019 = dataframes[3]
PIT_2019.head()
gender_nine = PIT_2019[["State","Overall Homeless - Female, 2019","Overall Homeless - Male, 2019","Overall Homeless - Transgender, 2019","Overall Homeless - Gender Non-Conforming, 2019"]]
gender_nine = gender_nine.replace(r'^\s*$', np.nan, regex=True)
gender_nine = gender_nine[["Overall Homeless - Female, 2019","Overall Homeless - Transgender, 2019","Overall Homeless - Male, 2019","Overall Homeless - Gender Non-Conforming, 2019"]].astype(float)
gender_nine.head()
gender_nine["State"] = PIT_2019["State"]
gender_nine = gender_nine.set_index("State")
sum_nine = gender_nine.loc["Total", :]
gender_nine_no_total = gender_nine.drop("Total")
gen_nin = gender_nine_no_total.plot.bar(stacked=True, title = 'Homeless Population in 2019, By Gender')
gen_nin.set_xlabel("State")
gen_nin.set_ylabel("Number of Homeless People")
```

```python
PIT_2020 = dataframes[4]
PIT_2020.head()
gender_twenty = PIT_2020[["State","Overall Homeless - Female, 2020","Overall Homeless - Male, 2020","Overall Homeless - Transgender, 2020","Overall Homeless - Gender Non-Conforming, 2020"]]
gender_twenty = gender_twenty.replace(r'^\s*$', np.nan, regex=True)
gender_twenty = gender_twenty[["Overall Homeless - Female, 2020","Overall Homeless - Transgender, 2020","Overall Homeless - Male, 2020","Overall Homeless - Gender Non-Conforming, 2020"]].astype(float)
gender_twenty["State"] = PIT_2020["State"]
gender_twenty = gender_twenty.set_index("State")
sum_twenty = gender_twenty.loc["Total", :]
gender_twenty_no_total = gender_twenty.drop("Total")
gen_twe = gender_twenty_no_total.plot.bar(stacked=True, title = 'Homeless Population in 2020, By Gender')
gen_twe.set_xlabel("State")
gen_twe.set_ylabel("Number of Homeless People")
```

```python
years = ["2017", "2018", "2019", "2020"]
fem_through_years = [sum_seven["Overall Homeless - Female, 2017"],sum_eight["Overall Homeless - Female, 2018"], sum_nine["Overall Homeless - Female, 2019"], sum_twenty["Overall Homeless - Female, 2020"]]
men_through_years = [sum_seven["Overall Homeless - Male, 2017"],sum_eight["Overall Homeless - Male, 2018"], sum_nine["Overall Homeless - Male, 2019"], sum_twenty["Overall Homeless - Male, 2020"]]
trans_through_years = [sum_seven["Overall Homeless - Transgender, 2017"],sum_eight["Overall Homeless - Transgender, 2018"], sum_nine["Overall Homeless - Transgender, 2019"], sum_twenty["Overall Homeless - Transgender, 2020"]]
nonconform_through_years = [sum_seven["Overall Homeless - Gender Non-Conforming, 2017"],sum_eight["Overall Homeless - Gender Non-Conforming, 2018"], sum_nine["Overall Homeless - Gender Non-Conforming, 2019"], sum_twenty["Overall Homeless - Gender Non-Conforming, 2020"]]
gender_total_throughout_years = pd.DataFrame({'Female':fem_through_years, 'Male':men_through_years, 'Transgender':trans_through_years, 'Gender Non-Conforming': nonconform_through_years}, index=years)
gender_total = gender_total_throughout_years.plot.line(title = "Homeless Population from 2017 to 2020")
gender_total.set_xlabel("Years")
gender_total.set_ylabel("Number of Homeless People")
gender_total.grid()
```
